```python
import pandas as pd

data = pd.read_csv("sadboihours.csv", sep=",")
data.sample(10)
energy = data["energy"].mean()
dance = data["danceability"].mean()
speech = data["speechiness"].mean()
live = data["liveness"].mean()
valence = data["valence"].mean()
print("Energy = ",energy)
print("Danceability = ",dance)
print("Speechiness = ",speech)
print("Liveness = ",live)
print("Valence = ",valence)
```

```python
import pandas as pd
chip = pd.read_csv("sadboihours.csv", sep=",")
values = chip.query('energy <= 0.507 & danceability <= 0.59 & speechiness <= 0.07 & liveness <= 0.17 & valence <= 0.31')
values.head(20)
```
